The goal here is to incorporate both sessions and users data into one dataset. There are several obstacles:

* <b>Age</b> column has various outliers e.g 2014, 1, 1956, etc. Solution: leave only valid values and categorize the other into '0-14', '100-1900', '1900-2014'
* <b>Dates</b> are needed to be broken into year, month, day, season, weekday, difference and so on.
* Categorical data should be one-hot encoded, to reduce the feature space mostly only 5-10 most common categories considered
* In sessions we need to group everything by id, aggregating with different functions to produce new features: number of actions, common unique actions, mean time elapsed, device types and so on
* NaNs. In sessions we just do not count it. In users fill them with -1

```python
action_count = groupby.count()['action']
action_unique = groupby['action'].unique()
action_type_unique = groupby['action_type'].unique()
action_detail_unique = groupby['action_detail'].unique()
device_type_unique = groupby['device_type'].unique()

for key in action_unique.keys():
    action_unique[key] = len(action_unique[key])
    action_type_unique[key] = len(action_type_unique[key])
    action_detail_unique[key] = len(action_detail_unique[key])
    device_type_unique[key] = len(device_type_unique[key])
```

```python
devices = ['Mac Desktop', 'Windows Desktop', 'iPhone', 'Android Phone', 'iPad Tablet', 'Other']
actions = ['show', 'index', 'search_results', 'personalize', 'search', 'ajax_refresh_subtotal', 'update', 'similar_listings', 
           'social_connections', 'reviews', 'other']
actions_type = ['view', 'data', 'click', '-unknown-', 'submit', 'message_post','partner_callback', 'booking_request', 'modify',
                'booking_response']
```

```python
d_devices = {}
d_actions = {}
d_actions_type = {}
count =0
for g in groupby:
    top = g[1]['action'].value_counts().keys()
    common = set(top[:min(10,len(top))])&set(actions)
    bin0 = []
    for action in actions:
        if action in common:
            bin0.append(1)
        else:
            bin0.append(0)
    d_actions[g[0]] = bin0
    #print actions_bin
    top = g[1]['action_type'].value_counts().keys()
    common = set(top[:min(10,len(top))])&set(actions_type)
    bin0 = []
    for action in actions_type:
        if action in common:
            bin0.append(1)
        else:
            bin0.append(0)
    d_actions_type[g[0]] = bin0
    top = g[1]['device_type'].value_counts().keys()
    common = set(top[:min(6,len(top))])&set(devices)
    bin0 = []
    for device in devices:
        if device in common:
            bin0.append(1)
        else:
            bin0.append(0)
    d_devices[g[0]] = bin0
```

```python
df_devices = pd.DataFrame(columns=devices)
df_devices = df_devices.from_dict(d_devices, orient='index')
df_devices.columns = devices
df_actions_type = pd.DataFrame(columns=actions_type)
df_actions_type = df_actions_type.from_dict(d_actions_type, orient='index')
df_actions_type.columns = actions_type
df_actions = pd.DataFrame(columns=actions)
df_actions = df_actions.from_dict(d_actions, orient='index')
df_actions.columns = actions
```

```python
secs_elapsed_sum = groupby['secs_elapsed'].sum()
secs_elapsed_mean = groupby['secs_elapsed'].mean()
secs_elapsed_min = groupby['secs_elapsed'].min()
secs_elapsed_max = groupby['secs_elapsed'].max()
```

```python
df_actions['secs_elapsed_sum'] = secs_elapsed_sum
df_actions['secs_elapsed_min'] = secs_elapsed_min
df_actions['secs_elapsed_max'] = secs_elapsed_max
df_actions['secs_elapsed_mean'] = secs_elapsed_mean 
df_actions['device_type_unique'] = device_type_unique
df_actions['action_type_unique'] = action_type_unique
df_actions['action_detail_unique'] = action_detail_unique
df_actions['action_unique'] = action_unique
df_actions['action_count'] = action_count
df_actions['id'] = df_actions.index
df_actions_type['id'] = df_actions_type.index
df_devices['id'] = df_devices.index
```

```python
df_sessions = pd.merge(df_devices, df_actions, on='id')
df_sessions = pd.merge(df_sessions, df_actions_type, on='id')
```

```python
date_account_created = train['date_account_created']
timestamp_first_active = train['timestamp_first_active']
ids = train['id']
```

```python
import datetime
import calendar
dates = {}
dates_header = ['id','acc_year','acc_month','acc_day','stamp_year','stamp_month','stamp_day','stamp_hour','acc_weekday',
                'stamp_weekday','acc_season','stamp_season','date_diff']
seasons = ['winter', 'winter', 'spring', 'spring', 'spring', 'summer', 'summer', 'summer', 'autumn', 'autumn', 'autumn', 'winter']
for key in timestamp_first_active.keys():
    acc_year = int(date_account_created[key][0:4])
    acc_month = int(date_account_created[key][5:7])
    acc_day = int(date_account_created[key][8:10])
    stamp_year = int(str(timestamp_first_active[key])[0:4])
    stamp_month = int(str(timestamp_first_active[key])[4:6])
    stamp_day = int(str(timestamp_first_active[key])[6:8])
    stamp_hour = int(str(timestamp_first_active[key])[8:10])
    date1 = datetime.date(acc_year, acc_month, acc_day)
    acc_weekday = calendar.day_name[date1.weekday()]
    date2 = datetime.date(stamp_year, stamp_month, stamp_day)
    stamp_weekday = calendar.day_name[date2.weekday()]
    acc_season = seasons[acc_month-1]
    stamp_season = seasons[stamp_month-1]
    date_diff = (date1-date2).days
    dates[key] = [ids[key],acc_year,acc_month,acc_day,stamp_year,stamp_month,stamp_day,stamp_hour,acc_weekday,stamp_weekday,
                  acc_season,stamp_season,date_diff]
df_dates = pd.DataFrame(columns=dates_header)
df_dates = df_dates.from_dict(dates, orient='index')
df_dates.columns = dates_header
```

```python
ages = train['age']
```

```python
d_ages = {}
ages_header = ['id', 'normal_age', '0-14', '100-1900', '1900-2014']
for key in ages.keys():
    arr = [0]*5
    if not pd.isnull(ages[key]):
        if ages[key]>=1900:
            arr[4] = 1
        if (ages[key]<1900) and (ages[key]>=100):
            arr[3] = 1  
        if (ages[key]<14) and (ages[key]>=0):
            arr[2] = 1 
        if (ages[key]<100) and (ages[key]>=14):
            arr[1] = ages[key] 
    else:
        arr = [-1]*5
    arr[0] = ids[key]
    d_ages[key] = arr
df_ages = pd.DataFrame(columns=ages_header)
df_ages = df_ages.from_dict(d_ages, orient='index')
df_ages.columns = ages_header
```

```python
browser = []
allowed = set(['Chrome','Safari','Firefox','-unknown-','IE','Mobile Safari'])
for v in train['first_browser']:
    if v in allowed:
        browser.append(v)
    else:
        browser.append('other')
```
